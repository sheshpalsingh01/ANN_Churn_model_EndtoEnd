```python
## Load Lib
import pickle
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import load_model
```

```python
# Load all file
with open('01_encoded_gender.pkl','rb') as f:
    encoded_gender=pickle.load(f)
with open('02_encoded_geo.pkl','rb') as f:
    encoded_geo= pickle.load(f)
with open('03_scaler.pkl','rb') as f:
    scaler=pickle.load(f)
ANNmodel=load_model('04_model.h5')
```

```python
input_data ={
    'CreditScore':600, 
    'Gender':'Male', 
    'Age':40, 
    'Tenure':3, 
    'Balance':60000, 
    'NumOfProducts':4,
    'HasCrCard':1, 
    'IsActiveMember':1, 
    'EstimatedSalary':5000, 
    'Geography':'France',
    
}
```

```python
# Geo Encoded
geo_encoded=encoded_geo.transform([[input_data['Geography']]]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded,columns=encoded_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
input_df=pd.DataFrame([input_data])
input_df=pd.concat([input_df.reset_index(drop=True),geo_encoded_df],axis=1)
input_df
```

```python
## Gender encoding
input_df['Gender'] =encoded_gender.fit_transform(input_df['Gender'])
input_df
```

```python
input_df=input_df.drop('Geography',axis=1)
input_df
```

```python
## Scaling
scaler=scaler.transform(input_df)
scaler
```

```python
prediction=ANNmodel.predict(input_df)
prediction
```
